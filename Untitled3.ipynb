```python
import torch
from diffusers import StableDiffusionPipeline

# Load the Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Use GPU for faster processing
```

```python
# Define your prompt
prompt = "A vibrant yellow bird with orange cheek patches perched on a wooden branch, with a soft blurred green forest background. The bird has a bright crest, detailed feathers, and a curious expression. Hyper-realistic, 4K, high detail, professional photography lighting"

import matplotlib.pyplot as plt

# Generate the image
image = pipe(prompt).images[0]

# Save and display using Matplotlib
plt.imshow(image)
plt.axis("off")  # Hide axes
plt.show()
```
